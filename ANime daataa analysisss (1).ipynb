```python
x=df[['members','duration']]
y=df['score']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python

y_pred = reg.predict(x_test)
pred=reg.predict(x)
```

```python
sc=reg.score(x_test,y_test)
sc
```

```python
plt.figure(figsize=(16,8))
plt.plot(x_test,y_test, 'ob', label = "Actual")
plt.plot(x_test,y_pred, 'r', label = "Predicted")
plt.ylabel('Modal Price', rotation = 0, fontsize = 14)
plt.xlabel('Month', fontsize = 14)
plt.legend(loc='top right')
plt.grid()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_forest = RandomForestRegressor()
regressor_forest.fit(x_train, y_train)
```

```python
from sklearn.svm import SVR
regressor_svr = SVR(kernel = 'rbf')
regressor_svr.fit(x_train,y_train)
```

```python
y_pred_svr=regressor_svr.predict(x_test)
```

```python
scr_svr=regressor_svr.score(x_test,y_test)
scr_svr
```
